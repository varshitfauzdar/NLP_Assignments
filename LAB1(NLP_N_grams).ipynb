```python
import nltk
import spacy
import string
import random
from collections import Counter, defaultdict
from textblob import TextBlob

# Download NLTK resources
nltk.download('punkt')
nltk.download('gutenberg')
# Load English language model
!python -m spacy download en_core_web_sm
nlp = spacy.load('en_core_web_sm')
```

```python
from nltk.tokenize import word_tokenize

nltk.download('punkt_tab')

sentence = "The quick brown fox jumps over the lazy dog."
tokens = word_tokenize(sentence)

print("Tokens:", tokens)
```

```python
def clean_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text


print(clean_text(sentence))
```

```python
import wikipedia

pages = [
    "Natural language processing",
    "Artificial intelligence",
    "Machine learning",
    "Deep learning",
    "Computer science"
]

wiki_text = ""

for page in pages:
        wiki_text += wikipedia.page(page, auto_suggest=False).content
        print(f"Downloaded: {page}")

# print(wiki_text)
```

```python
wiki_text = clean_text(wiki_text)
words = wiki_text.split()

print("Total words:", len(words))
```

```python
unigram_counts = Counter(words)
total_words = len(words)

def unigram_prob(word):
    return unigram_counts[word] / total_words
```

```python
bigram_counts = defaultdict(Counter)

for i in range(len(words) - 1):
    bigram_counts[words[i]][words[i+1]] += 1

def bigram_prob(w1, w2):
    return bigram_counts[w1][w2] / sum(bigram_counts[w1].values())
```

```python
def generate_sentence(start_word, length=15):
    sentence = [start_word]

    for _ in range(length - 1):
        current_word = sentence[-1]

        if current_word in bigram_counts and bigram_counts[current_word]:
            next_word = max(
                bigram_counts[current_word],
                key=lambda w: bigram_prob(current_word, w)
            )
        else:
            next_word = random.choice(list(unigram_counts.keys()))

        sentence.append(next_word)

    return " ".join(sentence)

start_word = "computer"
generated_sentence = generate_sentence(start_word, 15)
print("Generated Sentence:")
print(generated_sentence)
```
